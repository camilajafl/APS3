```python
import numpy as np
from math import pi, sin, cos, sqrt
import matplotlib.pyplot as plt



def força_A(theta):

    m=42.86399
    g=8.91
    r_barra =0.303328
    Ig=1.43313

    alfa = ((m*g*r_barra)*cos(theta))/(m*r_barra + Ig)

    Ft = -Ig*alfa/r_barra
    Fn = m*g*sin(theta)
    Fa = sqrt(Ft**2 + Fn**2)

    return Fa

angulo = np.arange(0, pi + 0.01, 0.01)

resultados = []

for theta in angulo:
    forca = força_A(theta)
    resultados.append(forca)


plt.plot(angulo,resultados )  
plt.xlabel('theta')
plt.ylabel("força")
plt.title('força x theta')
plt.grid()
plt.show()

```
